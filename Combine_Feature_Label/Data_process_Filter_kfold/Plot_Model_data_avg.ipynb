```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Combined_model_performance_results.csv"
df = pd.read_csv(file_path)

# Extract unique metrics excluding 'Model' and 'Dataset'
metrics = df.columns[2:]

# Convert metric values to float (removing ± and standard deviation)
for metric in metrics:
    df[metric] = df[metric].str.split('±').str[0].astype(float)

# Generate plots for each metric
for metric in metrics:
    plt.figure(figsize=(10, 5))
    
    # Pivot data for better plotting
    pivot_df = df.pivot(index="Dataset", columns="Model", values=metric)
    
    # Plot bars
    pivot_df.plot(kind="bar", figsize=(10, 5))
    plt.title(f"Model {metric.replace('_', ' ').title()} Comparison Across Datasets")
    plt.ylabel(metric.replace('_', ' ').title())
    plt.xlabel("Dataset")
    plt.xticks(rotation=0)
    plt.legend(title="Model")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    
    # Show the plot
    plt.show()
```
